### Implementations

#### Gradient Descent Implementation

```python
def GD_single_var(X0, f, Df, DDf, LR, Epsilon):
    # step_1
    x_0 = X0
    i = 0
         
    while True:
    # step_2
        d_x_0 = Df(x_0)
   
        if (d_x_0 < Epsilon):       
            break
            
        i+=1
    # step_3
        x_0 = x_0 - (LR * d_x_0)
        
    print(f"GD Result with x0 = {X0}, lr = {LR}, epsilon = {Epsilon}:\n")
    print(f"Found solution after {i} iterations")
    print(f"x_min = {x_0}")
    print(f"Gradient = {d_x_0}")
```

#### Newton's Method implementation

```python
def Newton_single_var(X0, f, Df, DDf, LR, Epsilon):
    # step_1:
    x_0 = X0
    i = 0
    
    while True:
    # step_2:
        d_x_0 = Df(x_0)
        dd_x_0 = DDf(x_0)
    
        if (d_x_0 < Epsilon):       
            break
        
        i+=1
    
    # step_3:
        x_0 = x_0 - (LR * (d_x_0 / dd_x_0)) 
        
    print(f"Newton Result with x0 = {X0}, lr = {LR}, epsiln = {Epsilon}:\n")
    print(f"Found solution after {i} iterations")
    print(f"x_min = {x_0}")
    print(f"Gradient = {d_x_0}")
        
    return i, x_0, d_x_0
```

### Implementations

#### GD implementation

```python
def GD_Multi_var(X, f, Df, DDf, LR, Epsilon):
    # step_1:
    x = X
    i = 0
    
    while True:
    # step_2:
        d_x = Df(x)
        dd_x = DDf(x)
        d_x_norm = np.linalg.norm(d_x)
    
        if (d_x_norm < Epsilon):       
            break
        
        i+=1
    
    # step_3:
        x = x - (LR * d_x) 
        
    print(f"GD Result with x = {X}, lr = {LR}, epsilon = {Epsilon}:\n")
    print(f"Found solution after {i} iterations")
    print(f"x_min = {x}")
    print(f"Gradient = {d_x}")
        
```

#### Newton implementation

```python
def Newton_Multi_var(X, f, Df, DDf, LR, Epsilon):
    # step_1:
    x = X
    i = 0
    
    while True:
    # step_2:
        d_x = Df(x)
        dd_x = DDf(x)
        H_inv = np.linalg.inv(dd_x)
        d_x_norm = np.linalg.norm(d_x)
    
        if (d_x_norm < Epsilon):       
            break
        
        i+=1
    
    # step_3:
        x = x - (LR * ((H_inv) @ d_x)) 
        
    print(f"Newton Result with x = {X}, lr = {LR}, epsilon = {Epsilon}:\n")
    print(f"Found solution after {i} iterations")
    print(f"x_min = {x}")
    print(f"Gradient = {d_x}")
```

## BFGS
- Implement the <b>BFGS</b> as a function and find the minimum of the following function:
    - $ùíá(ùíô)=ùíô^ùüê+y^2$
    - $ùíá(ùíô)=ùíô^ùüê+xy+y^2$  
    - $ùíá(ùíô)=ùíô^ùüêy$ 
- The stop condition is grad < epsilon.
- Try different epsilons and initial conditions.
- Compare the results from Gradient descent and Newton's method.

```python
def BFGS_Quasi_Method(X_prev, X_curr, f, Df, DDf, LR, Epsilon):
# step_1:
    x_prev = X_prev
    x_curr = X_curr
    B_inv = np.identity(len(x_curr))
    iden_mat = np.identity(len(x_curr))
    i = 0
    
    while True:
    # step_2:
        d_x_prev = Df(x_prev)
        d_x_curr = Df(x_curr)
        
        diff_X = x_curr - x_prev            # dx
        diff_d_X = d_x_curr - d_x_prev      # dy
        diff_x_curr_norm = np.linalg.norm(d_x_curr)
        
        print(f"\nxn_1:\n{x_prev}")
        print(f"\nxn:\n{x_curr}")
        print(f"\nDfxn_1:\n{d_x_prev}")
        print(f"\nDfxn:\n{d_x_curr}")
        print(f"\nNorm: {diff_x_curr_norm}\n")
        print("********************")
        
        if(diff_x_curr_norm < Epsilon):
            break
            
        i +=1

        B_inv = ((iden_mat - ((diff_X@diff_d_X.T)/(diff_d_X.T@diff_X))) @ B_inv @(iden_mat - ((diff_d_X@diff_X.T)/(diff_d_X.T@diff_X)))) + ((diff_X@diff_X.T)/(diff_d_X.T@diff_X))
        
        x_prev = x_curr
        x_curr = x_curr - (LR * (B_inv @ d_x_curr))
        
        print(f"B_inv:\n{B_inv}")
        print("*******************")
        
    print(f"\nFound Solution after {i} iterations")
    print(f"x0x1_min:\n{x_curr}")
    print(f"Gradient:\n{d_x_curr}")
```
